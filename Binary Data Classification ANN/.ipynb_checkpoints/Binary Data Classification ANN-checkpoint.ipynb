```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data preprocess
df = pd.read_csv(r'C:\Users\Hamza\Churn_Modelling.csv')
```

```python
df.head()
```

```python
#selecting dependent and independent
x = df.drop(labels=['RowNumber','CustomerId','Surname','Exited'],axis=1) #these are independent columns
y = df['Exited']
```

```python
x.head()
```

```python
y.head()
```

```python
#encoding catigoral data
from sklearn.preprocessing import LabelEncoder
```

```python
label_1 = LabelEncoder()
```

```python
x['Geography'] = label_1.fit_transform(x['Geography'])
```

```python
x.head()
```

```python
label_2 = LabelEncoder()
```

```python
x['Gender'] = label_1.fit_transform(x['Gender'])
```

```python
x.head(7)
```

```python
#Avoiding the dummpy variable to trap
x = pd.get_dummies(x,drop_first=True,columns=['Geography'])
```

```python
x.head(7)
```

```python
#split data into train and test set
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#feature sacling
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
#Buliding the model
model = tf.keras.models.Sequential()
```

```python
#adding input layers aand first hidden layer
#no of nodes = 6 units
#activation function = relu
#input dimention = 11 

model.add(tf.keras.layers.Dense(units=6,activation='relu', input_dim = 11))
```

```python
#add 2nd hidden layer
model.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
#output layers
model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

```python
#Complie the model
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
model.summary()
```

```python
#train the model
model.fit(x_train,y_train.to_numpy(), batch_size=10,epochs=20)
```

```python
# model eveloatiion and prediction
test_loss , test_acc = model.evaluate(x_test,y_test.to_numpy())
```

```python
print('Test Accuracy : {}'.format(test_acc))
```

```python

# Assuming you have imported the necessary libraries and defined your model

# Make predictions
y_pred_probs = model.predict(x_test)
y_pred = np.argmax(y_pred_probs, axis=1)
```
